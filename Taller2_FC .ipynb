```python
# Costos variables del transporte
c6 = np.array([[4,5,6,8,10],
              [6,4,3,5,8],
              [9,7,4,3,4]])

# Capacidad Maxima para almacenar
b6 = np.array([500,500,500])

# Costos fijos por cada planta
cf = np.array([1000,1000,1000])

# Demanda
b66 = np.array([80,270,250,160,180])

# variables
x6 = cp.Variable((3,5), nonneg = True)
# Variable de cerrar o abrir una planta
y6 = cp.Variable(3,boolean = True)
# Minimización
objetivo6 = cp.Minimize(cp.sum(cp.multiply(c6,x6))+cp.sum(cp.multiply(cf,y6)))

res6 = [cp.sum(x6[0,:])<=500*y6[0],cp.sum(x6[1,:])<=500*y6[1],
        cp.sum(x6[2:])<=500*y6[2],cp.sum(x6[:,0])==b66[0], 
        cp.sum(x6[:,1])==b66[1],cp.sum(x6[:,2])==b66[2],
        cp.sum(x6[:,3])==b66[3],cp.sum(x6[:,4])==b66[4],
        cp.sum(y6)==2]

prob5 = cp.Problem(objetivo6,res6)

result6 = prob5.solve()

data6 = pd.DataFrame(x6.value)

data6
```
